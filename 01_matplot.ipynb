### 绘图的操作

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10 , 100)
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x, siny, color = 'r', label = 'siny')
plt.plot(x, cosy, color = 'b', label = 'cosy')
plt.title("welcome to ML")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
```
